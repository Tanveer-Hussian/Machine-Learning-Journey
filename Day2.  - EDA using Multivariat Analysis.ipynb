```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
titanic = pd.read_csv('train.csv')
```

```python
titanic.head()
```

# Task 1: Categorical × Categorical
# Compare Survival rates across Sex and Pclass.

```python
sns.barplot(x=titanic['Survived'], y=titanic['Age'])
```

```python
sns.barplot(y=titanic['Age'], x=titanic['Survived'], hue=titanic['Pclass'])
```

# Task 2: Categorical × Numerical

# Analyze how Fare distribution varies across Pclass, separated by Survival.

```python
sns.boxplot(x=titanic['Pclass'], y=titanic['Fare'], hue=titanic['Survived'])
```

# Task 3: Numerical × Numerical × Categorical

# Plot Age vs Fare, color points by Survival, shape them by Pclass.

```python

sns.scatterplot(x=titanic['Age'], y=titanic['Fare'], hue=titanic['Survived'], style=titanic['Pclass'])
plt.title("Age vs Fare by Survival & Passenger Class")
plt.show()
```

# Task 4: Derived Feature × Numerical × Target

# Create a new feature: family_size = sibsp + parch + 1.

**See how Family Size relates to Survival, grouped by Sex.**

```python
titanic['family_size'] = titanic['SibSp']+titanic['Parch']+1
```

```python
titanic['family_size'].head()
```

```python
sns.barplot(x=titanic['family_size'], y=titanic['Survived'], hue=titanic['Sex'])
```

# Task 5: Heatmap Analysis

# Create two correlation heatmaps of numerical variables (age, fare, sibsp, parch, family_size ):

**One for Survivors only**

**One for Non-survivors**
  

```python
sns.heatmap(titanic[titanic['Survived']==1][['Age','Fare','SibSp','Parch','family_size']].corr())
```

```python
sns.heatmap(titanic[titanic['Survived']==0][['Age','Fare','SibSp','Parch','family_size']].corr())
```

# Task 6: Pairwise Visualization

# Use a pairplot (age, fare, family_size) with hue="survived" and kind="scatter".

```python
sns.pairplot(titanic[['Age','Fare','family_size','Survived']], hue='Survived')
plt.show()
```
