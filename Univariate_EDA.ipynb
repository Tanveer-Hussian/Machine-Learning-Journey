```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

# 1. Data Inspection

**Check data types (dtype) of each column**

**Count missing values per column**

**Identify categorical vs. numerical columns**

```python
# Check data types (dtype) of each column
columnList = ['PassengerId','Survived','Pclass','Name','Sex','Age',	'SibSp','Parch','Ticket','Fare','Cabin','Embarked']
for i in columnList:
    print(f"Data type of {i} is : {np.dtype(df[i])}")
```

```python
# Count missing values per column
for i in columnList: 
    print(f'Missing values in Column {i} are: {df[i].isnull().sum()}')
```

```python
# Identify categorical vs. numerical columns'
numericalCols = []
categoricalCols = []
for i in columnList: 
    if(np.dtype(df[i])== 'O'): 
        categoricalCols.append(i)
        print(f'{i} is categorical column')
    else: 
        numericalCols.append(i)
        print(f'{i} is numerical column')
```

```python
print(numericalCols)
print(categoricalCols)
```

# 2. Summary Statistics

**For numerical variables: mean, median, mode, min, max, standard deviation, quartiles**

**For categorical variables: frequency count, proportions**

```python

for i in numericalCols: 
    print(f'Mean of {i} is: {df[i].mean()}')
    print(f'Median of {i} is: {df[i].median()}')
    print(f'Mode of {i} is: {df[i].mode()[0]}')
    print(f'Min of {i} is: {df[i].min()}')
    print(f'Max of {i} is: {df[i].max()}') 
    print(f'Standard Deviation of {i} is: {df[i].std()}')
    print(f'Quartiles of {i} are: {df[i].quantile([0.25, 0.5, 0.75])}')
    
```

```python
# For categorical variables: frequency count, proportions
# For categorical variables: frequency count, proportions
print('frequency count of categorical Columns')
for i in categoricalCols: 
    print(f'Frequency count of {i} is: {df[i].value_counts()}') 

print('proportions of categorical Columns')
    print(f'proportion of {i} is: {df[i].value_counts(normalize=True)*100}') 
```

# 3. Distribution Analysis

**Plot histogram (numeric data).**

**Plot density/Kernel Density Estimation (KDE) for numeric features.**


```python
count = 0
plt.figure(figsize=(14,14))

for i in numericalCols:
    count = count+1
    plt.subplot(4,2,count)
    plt.hist(df[i])
    plt.title(df[i].name)
    
plt.tight_layout()
plt.show()
```

```python
count = 0
plt.figure(figsize=(14,14))
for i in numericalCols:
    count = count+1
    plt.subplot(4,2,count)
    sns.histplot(df[i], kde=True)
    plt.title(df[i].name)
       
plt.tight_layout() 
plt.show()
 
```
